importing librairies & methods 

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
from urllib.request import urlopen
```

inputting the URL

```python
url='https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=illustrator'
```

Create a client-based Request to get the URL

```python
client=urlopen(url)
```

Getting the HTML code of the Full page 

```python
html=client.read()
```

```python
html
```

Closing the Request

```python
client.close()
```

Creating an HTML Parser Using BeautifulSoup

```python
soup=bs(html, "html.parser")
```

```python
soup
```

Creating a Container for the Needed Data 

```python
containers= soup.find_all("div",{"class":"css-1gatmva e1v1l3u10"})
```

Bringing it All Together 

```python
f= open("job-illustor.csv", "w")
header= "job_title, company_name, job_type\n"
f.write(header)
```

```python
for container in containers:
    jtitle=container.find_all("h2",{"class":"css-m604qf"})
    job_title=jtitle[0].text.strip()
    cname=container.find_all("a",{"class":"css-17s97q8"})
    company_name=cname[0].text.strip()
    jtype=container.find_all("div",{"class":"css-1lh32fc"})
    job_type=jtype[0].text.strip()
    
    #print(job_title)
    #print(company_name)
    #print(job_type)
    #print()
    #print(job_title +", " + company_name +", "+ job_type)
    #print()
    f.write(job_title + ", " + company_name + ", " + job_type + "\n")
f.close()
```
